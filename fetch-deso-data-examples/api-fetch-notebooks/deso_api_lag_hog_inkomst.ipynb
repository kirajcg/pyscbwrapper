```python
from pyscbwrapper import SCB
```

```python
scb = SCB('sv', 'HE', 'HE0110', 'HE0110I', 'Tab4InkDesoN1')
scb.info()
```

```python
scb.get_variables()
```

```python
import re

regioner = scb.get_variables()['region']
r = re.compile('^0180C')
deso = list(filter(r.match, regioner))

scb.set_query(region=deso,
              ålder=["0-19 år",
                     "20-64 år",
                     "65+ år",
                     "totalt ålder"],
              tabellinnehåll=["Låg ekonomisk standard, procent",
                            "Hög ekonomisk standard, procent",
                            "Antal personer totalt"],
              år=["2020"])
```

```python
scb_data = scb.get_data()
```

```python
scb_data
```

```python
scb_uttag = scb_data['data']
scb_uttag
```

```python
koder = scb.get_query()['query'][0]['selection']['values']

desodic = {}

for i in range(len(koder)):
  desodic[koder[i]] = koder[i]

desodic
```

```python
desodata = {}
for kod in desodic:
    desodata[desodic[kod]] = {}
    
    for i in range(len(scb_uttag)):
        if scb_uttag[i]['key'][0] == kod:
            desodata[desodic[kod]][scb_uttag[i]['key'][1]] = scb_uttag[i]['values']
            
print(desodata)
```

```python
import pandas as pd

df1 = pd.DataFrame.from_dict(desodata, orient='index').rename_axis('deso').reset_index()
df1.rename(columns={'0-19':'z_n','20-64':'t_s','65+':'sfp'}, inplace=True)
df1
```

```python
df1[['0-19_low_pct','0-19_high_pct', '0-19_tot']] = pd.DataFrame(df1.z_n.tolist(), index= df1.index)
df1[['20-64_low_pct','20-64_high_pct', '20-64_tot']] = pd.DataFrame(df1.t_s.tolist(), index= df1.index)
df1[['65_plus_low_pct','65_plus_high_pct', '65_plus_tot']] = pd.DataFrame(df1.sfp.tolist(), index= df1.index)
df1[['tot_low_pct','tot_high_pct', 'tot_tot']] = pd.DataFrame(df1.tot.tolist(), index= df1.index)
df1.drop(['z_n', 't_s', 'sfp', 'tot'], axis=1, inplace=True)
df1
```

```python
df1.to_csv('../csv-output/deso_stockholm_lag_hog_inkomst.csv')
```
